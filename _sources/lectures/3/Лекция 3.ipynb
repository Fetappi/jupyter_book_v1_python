# Генераторы списков (List comprehensions, списковые включения)

```python
[<способ формирования значения> for <переменная> in <итерируемый объект>]
```

````python
N = 6
a = [0] * N

for i in range(N):
    a[i] = i ** 2

print(a)
````

````python
# [<способ формирования значения> for <переменная> in <итерируемый объект>]
a = [x ** 2 for x in range(N)]
a
````

# Тернарный оператор

```python
<значение 1> if <условие> else <значение 2>
```

````python
a = 12
b = 7

if a > b:
    res = a
else:
    res = b

print(res)
````

````python
a = 12
b = 7

res = a if a > b else b
print(res)
````

````python
a = -12
b = -7
res = abs(a) if a > b else abs(b)
res
````

````python
res = print(a) if a > b else print(b)
print(res)
````

````python
s = 'python'
t = 'upper'
res = s.upper() if t == 'upper' else s
res
````

````python
print([1, 2, a if a < b else b, 4, 5])
print("a - " + ("четное" if a % 2 == 0 else "нечетное") + " число")
````

````python
print(max(1, 5, a if a > 0 else b, 4, 5))
````

````python
a = 2
b = 3
c = -4

d = (a if a > c else c) if a > b else (b if b > c else c)
print(d)
````

````python
lst = [5, 3, 0, -6, 8, 10, 1]
````

````python
def get_filter(a, filter=None):
    if filter is None:
        return a

    res = []
    for x in a:
        if filter(x):
            res.append(x)

    return res
````

````python
lst = [5, 3, 0, -6, 8, 10, 1]
r = get_filter(lst)
print(r)
````

````python
r = get_filter(lst, lambda p: p % 2 == 0)
r
````

````python
def temp_filt(p):
    return p % 2 == 0
r = get_filter(lst, temp_filt)
r
````

````python
r = get_filter(lst, lambda p: p > 0)
r
````

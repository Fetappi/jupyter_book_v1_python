# Списки
[Смотреть](https://fadeevlecturer.github.io/python_lectures/notebooks/python/numbers_and_lists.html#id7)

Первый контейнер, с которым мы познакомимся --- список. Список (`list`) - наиболее близкий аналог к массиву из `C/C++`, если брать в расчет только встроенные типы. Существенным отличием является возможность хранения в одном и том же списке объекты разных типов.

В `python` ряд контейнеров обладают общими чертами. `list` --- представитель группы контейнеров `Sequence Types` (последовательности). [Последовательность](https://docs.python.org/3/library/stdtypes.html#sequence-types-list-tuple-range) означает, что элементы контейнера пронумерованы и к ним можно обращаться по индексу ([zero-based](https://en.wikipedia.org/wiki/Zero-based_numbering), как и в `C/C++` ).

Другие представители последовательностей в `python`:

* кортежи `tuple`;
* диапазоны `range`;
* строки `str`;

Создать список можно, например, перечислив его элементы внутри квадратных скобок и разделив их запятыми.

# Цикл `for`

```python
for переменная in итерируемый_объект:
    первая инструкция тела цикла
    ...
    последняя инструкция тела цикла
```









````python
l = [-10, 5, 3, 4, 5]

for i in l:
    print(i**2)
````

````python
l = ['a', 'b', 'c', 'd', 'e']

for i in range(len(l)):
    l[i] = 1
    print(i, ':', l[i])

print('-'*40)

for i, val in enumerate(l):
    val = 1
    print(i, ':', val)

print(l)
````

Стандартный вопрос:
Что из себя представляет та или иная функция?


````python
def sum_of_squares(x, y):
    return x**2 + y**2
````

````python
print(sum_of_squares(3, 4))
````

````python
'ab' > 'aa'
````

````python
def my_max(array):
    m = array[0]
    for i in array:
        if i > m:
            m = i
    return m


print(my_max([1, 10, 50, 200]))
print(my_max([200, 50, 10, 1]))
print(my_max(['a', 'b', 'A', 'B']))
````

````python
ord(' ')
````

````python
def change_x(x):
    print(id(x))
    x = 1000
    print(id(x))
    return x

a = 500
print(id(a))
a = change_x(a)
print(a)
````

````python
def change_list(l):
    print(id(l))
    l.append(6)
    print(id(l))

old_list = [1, 2, 3, 4, 5]
print(id(old_list))
change_list(old_list)
print(old_list)
````

# Тернарный оператор

```python
<значение 1> if <условие> else <значение 2>
```

````python
a = 12
b = 7

if a > b:
    res = a
else:
    res = b

print(res)
````

````python
a = 12
b = 7

res = a if a > b else b
# res = a > b ? a : b;
print(res)
````

````python
a = -12
b = -7
res = abs(a) if a > b else abs(b)
res
````

````python
res = print(a) if a > b else print(b)
print(res)
````

````python
s = 'python'
t = 'upper'
res = s.upper() if t == 'upper' else s
res
````

````python
print([1, 2, a if a < b else b, 4, 5])
print("a - " + ("четное" if a % 2 == 0 else "нечетное") + " число")
````

````python
print(max(1, 5, a if a > 0 else b, 4, 5))
````

````python
a = 2
b = 3
c = -4

d = (a if a > c else c) if a > b else (b if b > c else c)
print(d)
````

````python
d = [4, 3, -5, 0, 2, 11, 122, -8, 9]
a = ["четное" if x % 2 == 0 else "нечетное" for x in d]
a
````

````python
a = ["четное" if x % 2 == 0 else "нечетное"
     for x in d
     if x > 0
]
a
````

# Вложенные генераторы списков

````python
a = [(i, j) for i in range(3) for j in range(4)]
print(a)
````

````python
a = [(i, j)
     for i in range(3)
     for j in range(4)
]
print(a)
````

````python
a = [(i, j)
     for i in range(3) if i % 3 == 0
     for j in range(4)
]
print(a)
````

````python
a = [(i, j)
     for i in range(3) if i % 3 == 0
     for j in range(4) if j % 2 != 0
]
print(a)
````

````python
a = [f"{i}*{j} = {i*j}"
     for i in range(1, 4)
     for j in range(1, 4)
]
print(a)
````

````python
matrix = [[0, 1, 2, 3],
          [10, 11, 12, 13],
          [20, 21, 22, 23]]

a = [x
     for row in matrix
     for x in row
]
print(a)
````

````python
lst = [5, 3, 0, -6, 8, 10, 1]
````

````python
def get_filter(a, filter=None):
    if filter is None:
        return a

    res = []
    for x in a:
        if filter(x):
            res.append(x)

    return res
````

````python
lst = [5, 3, 0, -6, 8, 10, 1]
r = get_filter(lst)
print(r)
````

````python
r = get_filter(lst, lambda p: p % 2 == 0)
r
````

````python
def temp_filt(p):
    return p % 2 == 0
r = get_filter(lst, temp_filt)
r
````

````python
r = get_filter(lst, lambda p: p > 0)
r
````
